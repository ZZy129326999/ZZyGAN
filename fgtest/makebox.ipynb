```python
import os 
from PIL import Image

path = r'/stu09/faster-rcnn-pytorch/VOCdevkit/VOC2007/JPEGImages/'

f1 = open('/stu09/fgtest/data/chair_images3.txt', 'w')
f2 = open('/stu09/fgtest/data/chair_bounding_boxes3.txt', 'w')

i = 1
j = 0
for name in os.listdir(path):
    if 'jpg' not in name:
        continue
    image = Image.open(path + '/' + name)
    f1.write(str(i) + ' ')
    f1.write(str(j).zfill(3) + '.' + 'chair' + '/' + name + '\n')
    print(i, name)
    f2.write(str(i) + ' ')
    i += 1
    f2.write(str(0.0) + ' ' + str(0.0) + ' ' + str(image.size[0]) + ' ' + str(image.size[1]) + '\n')

f1.close()
f2.close()
```
